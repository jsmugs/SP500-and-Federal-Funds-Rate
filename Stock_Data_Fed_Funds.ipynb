```python
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
from matplotlib.dates import MonthLocator, DateFormatter
import matplotlib.dates as mdates


df = pd.read_csv('/content/drive/MyDrive/Raw Data.csv')

df['Date'] = pd.to_datetime(df['Date'])

x1 = df['Date']
y1 = df['Adj Close']
y2 = df['FEDFUNDS']

fig = plt.figure()

fig, ax1 = plt.subplots(figsize=(15, 7.5))
ax2 = ax1.twinx()
ax1.plot(x1, y1, 'b-', label='S&P500')
ax2.plot(x1,y2, 'g-', label='FEDFUNDS') 
#ax1.set_xlabel('Year')  
ax1.set_ylabel('Price[$]')
ax2.set_ylabel('Percent')  
ax1.set_title("S&P500 and Federal Funds Rate", fontdict={'fontname': 'DejaVu Sans','fontsize': 20})  


date_form = DateFormatter("%Y")
ax1.xaxis.set_major_formatter(date_form)
locator = mdates.YearLocator(2)
ax1.xaxis.set_major_locator(locator)

plt.savefig('S&P500_and_FEDFUNDS.png')
ax1.legend()
ax2.legend();
```
